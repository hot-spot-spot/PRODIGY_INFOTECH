### Extraction of the top 10 countries based on total population and creating bar charts on the years 2022, 2016, 2010 and 2001

```python
total_population_data = df[df['Series Code'] == 'SP.POP.TOTL']
total_population_sorted = total_population_data.sort_values(by="2022", ascending=False)
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population\n")
print(total_top_ten_countries[['Country Name']])
```

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Name", data=total_top_ten_countries, palette="deep", hue="Country Name", legend=False)
plt.title("Top 10 Countries of Total Population in 2022", fontsize=10)
plt.xlabel("Total Population", fontsize=10)
plt.ylabel("Country", fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2016", y="Country Name", data=total_top_ten_countries, palette="deep", hue="Country Name", legend=False)
plt.title("Top 10 Countries of Total Population in 2016", fontsize=10)
plt.xlabel("Total Population", fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2010", y="Country Name", data=total_top_ten_countries, palette="deep", hue="Country Name", legend=False)
plt.title("Top 10 Countries of Total Population in 2010", fontsize=10)
plt.xlabel("Total Population", fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2001", y="Country Name", data=total_top_ten_countries, palette="deep", hue="Country Name", legend=False)
plt.title("Top 10 Countries of Total Population in 2001", fontsize=10)
plt.xlabel("Total Population", fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```
